```python
# encontrando correlación, se encontró correlacion  mas fuerte entre:
# writing y reading 0.96 
# 0.83 entre math y reading
# 0.81 entre  writing y math
matriz=stuperf_df1[['math score', 'reading score', 'writing score']].corr()
```

```python
#esto es para visualizar la matriz de correlación con seaborn
#en donde se observa visualmente las correlaciones
import seaborn as sns
sns.heatmap(matriz, vmin = -1, vmax = +1, annot = True, cmap = 'bwr')
```
